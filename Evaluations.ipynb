# **Experimental Pipeline**


To evaluate the performance of the model using cosine similarity instead of L1 Norm in the point tracking process, the Image Fidelity and Mean distance metrics were calculated. Multiple hyperparameters differ from the original DragDiff results due to computing limitations. However, with the selected settings, the performance is still good enough to assess the proof-of-concept.

### 1. Unzip DragBench data into "drag_bench_data" folder
Only the categories of 'art_work', 'building_city_view', and 'human_full_body' were used.

### 2. Train LoRA with "run_lora_training.py"
Batch size was reduced from 4 to 2 to execute on the available GPU memory.

### 3. Run difussion model with "run_drag_diffusion.py" 
The number of fine-tuning steps (defined by *lora_steps*) were reduced from 80 to 40. With this number of steps, a decent performance is still achieved, as demonstrated in Figure 7(b) of the original DragDiffusion paper.

### 4. Run this notebook fpr evaluating Image Fidelity and Mean Distance

# **Plotting**

```python
import pandas as pd
import matplotlib.pyplot as plt

# LPIPS
lpips = {
    'Category': ['Art Work', 'Art Work', 'Art Work', 'City View', 'City View', 'City View', 'Full body human', 'Full body human', 'Full body human'],
    'Image': [1, 2, 3, 1, 2, 3, 1, 2, 3],
    'L1 Norm': [0.0521, 0.1578, 0.1758, 0.0974, 0.1514, 0.2501, 0.0508, 0.054, 0.0757],
    'cosine similarity': [0.0507, 0.198, 0.3736, 0.0984, 0.1512, 0.2299, 0.0513, 0.0836, 0.0788]
}

lpips_df = pd.DataFrame(lpips)

# 1 - LPIPS
image_fidelity_df = lpips_df.copy()
image_fidelity_df['L1 Norm'] = 1 - image_fidelity_df['L1 Norm']
image_fidelity_df['cosine similarity'] = 1 - image_fidelity_df['cosine similarity']


# Mean Distance
mean_distance_data = {
    'Category': ['Art Work', 'Art Work', 'Art Work', 'City View', 'City View', 'City View', 'Full body human', 'Full body human', 'Full body human'],
    'Image': [1, 2, 3, 1, 2, 3, 1, 2, 3],
    'L1 Norm': [6.0616, 42.7551, 27.6643, 22.3607, 82.0975, 54.8454, 2.8284, 40.537, 193.4942],
    'cosine similarity': [13.0178, 28.6356, 97.7901, 21.9317, 38.6005, 28.4122, 3.6056, 217.2664, 190.2761]
}

mean_distance_df = pd.DataFrame(mean_distance_data)


# ==================== PLOTTING ==================== #

# Calculate averages for Image Fidelity and Mean Distance
average_data = pd.DataFrame({
    'Category': ['Art Work', 'City View', 'Full body human'],
    'L1 Norm Mean Distance': mean_distance_df.groupby('Category')['L1 Norm'].mean().values,
    'Cosine Sim Mean Distance': mean_distance_df.groupby('Category')['cosine similarity'].mean().values,
    'L1 Norm Image Fidelity': image_fidelity_df.groupby('Category')['L1 Norm'].mean().values,
    'Cosine Sim Image Fidelity': image_fidelity_df.groupby('Category')['cosine similarity'].mean().values,
})

# Plot assets
plt.figure(figsize=(10, 6))
color_map = {'L1 Norm': 'blue', 'Cosine Similarity': 'orange'}
marker_map = {'Art Work': 'o', 'City View': 's', 'Full body human': '^'}

# Plot category and method
for _, row in average_data.iterrows():
    category = row['Category']
    # Plot L1 Norm
    plt.scatter(
        row['L1 Norm Mean Distance'], 
        row['L1 Norm Image Fidelity'],
        color=color_map['L1 Norm'],
        marker=marker_map[category],
        s=300,
        label=f"{category} - L1 Norm"
    )
    # Plot Cosine Similarity
    plt.scatter(
        row['Cosine Sim Mean Distance'], 
        row['Cosine Sim Image Fidelity'],
        color=color_map['Cosine Similarity'],
        marker=marker_map[category],
        s=300,
        label=f"{category} - Cosine Sim"
    )

# Plot assets
category_markers = [plt.Line2D([0], [0], marker=marker, color='w', label=cat, markerfacecolor='k', markersize=10)
                    for cat, marker in marker_map.items()]
method_colors = [plt.Line2D([0], [0], marker='o', color='w', label=method, markerfacecolor=color, markersize=10)
                 for method, color in color_map.items()]

plt.legend(handles=category_markers + method_colors, fontsize=16, loc='lower right', bbox_to_anchor=(1, 0))
plt.tick_params(axis='both', labelsize=16)
plt.xlabel("Mean Distance", fontsize=20)
plt.ylabel("Image Fidelity (1-LPIPS)", fontsize=20)
plt.grid(True)
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# LPIPS Data 
lpips_data = {
    'Category': ['Art Work', 'City View', 'Full body human'],
    'L1 Norm': [0.12855183333158493, 0.1663139263788859, 0.0601630707581838],
    'cosine similarity': [0.20744339873393378, 0.15984188516934714, 0.07123615841070811]
}
lpips_df = pd.DataFrame(lpips_data)

# 1 - LPIPS
image_fidelity_df = lpips_df.copy()
image_fidelity_df['L1 Norm'] = 1 - image_fidelity_df['L1 Norm']
image_fidelity_df['cosine similarity'] = 1 - image_fidelity_df['cosine similarity']


# Mean Distance Data (Averaged)
mean_distance_data = {
    'Category': ['Art Work', 'City View', 'Full body human'],
    'L1 Norm': [22.0414, 53.5372, 69.3491],
    'cosine similarity': [50.0503, 29.3391, 157.1036]
}
mean_distance_df = pd.DataFrame(mean_distance_data)

# Calculate global averages for Image Fidelity and Mean Distance
global_averages = {
    'Metric': ['L1 Norm', 'Cosine Similarity'],
    'Mean Distance': [
        mean_distance_df['L1 Norm'].mean(),
        mean_distance_df['cosine similarity'].mean()
    ],
    'Image Fidelity': [
        image_fidelity_df['L1 Norm'].mean(),
        image_fidelity_df['cosine similarity'].mean()
    ]
}

global_avg_df = pd.DataFrame(global_averages)

# ==================== PLOTTING ==================== #
plt.figure(figsize=(10, 6))

# Map colors for the two metrics
color_map = {'L1 Norm': 'blue', 'Cosine Similarity': 'orange'}

# Plot points for the two metrics
for _, row in global_avg_df.iterrows():
    plt.scatter(
        row['Mean Distance'],
        row['Image Fidelity'],
        color=color_map[row['Metric']],
        s=400,
        label=row['Metric']
    )

plt.legend(fontsize=16, loc='lower left')
plt.tick_params(axis='both', labelsize=16)
plt.xlabel("Mean Distance", fontsize=20)
plt.ylabel("Image Fidelity (1-LPIPS)", fontsize=20)
plt.grid(True)
plt.tight_layout()
plt.show()

```
